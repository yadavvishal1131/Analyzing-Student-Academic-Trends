> # ððžð«ðŸð¨ð«ð¦ðšð§ðœðž ð†ð«ð¨ð®ð©ð¢ð§ð 

```python
# Define performance based on exam_score
def performance_group(exam_score):
    if exam_score >= 45:
        return "Excellent"
    elif exam_score >= 35:
        return "Good"
    elif exam_score >= 25:
        return "Average"
    else:
        return "Poor"
    
```

```python
# Apply function to create new column

df['Performance_level'] = df['exam_score'].apply(performance_group)
print(df)
```

```python
group_countS = df['Performance_level'].value_counts()
print("\nNumber of student in each group:")
print(group_countS)
```

> # ð€ð¯ðžð«ðšð ðž ð„ð±ðšð¦ ð’ðœð¨ð«ðž ð›ð² ð€ð­ð­ðžð§ððšð§ðœðž

```python
# Define attendance bins
bins = [0,60,80,100]
labels = ['Low','Medium','High']
```

```python
# Create a new column for attendance category
df['attendance_group'] = pd.cut(df['attendance_percent'], bins=bins, labels=labels, right=True)
```

```python
# Group by attendance group and calculate average exam score
avg_score_by_attendance = df.groupby('attendance_group')['exam_score'].mean()
```

```python
print(avg_score_by_attendance)
```

```python
# Case 1: High study (>9 hours) and Low sleep (<6 hours)
high_study_low_sleep = df[(df['hours_studied']>9) & (df['sleep_hours']<6)]
print("High Study & Low Sleep Students:")
print(high_study_low_sleep)
```

```python
# Case 2: Low study (<9 hours) and High sleep (>6 hours)
low_study_high_sleep = df[(df['hours_studied'] < 9) & (df['sleep_hours'] > 6)]
print("\nLow Study & High Sleep Students:")
print(low_study_high_sleep)
```

```python
# Optional: Average exam score in both cases
avg_high_study_low_sleep = high_study_low_sleep['exam_score'].mean()
avg_low_study_high_sleep = low_study_high_sleep['exam_score'].mean()

print("\nAverage exam score for High Study & Low Sleep:", avg_high_study_low_sleep)
print("Average exam score for Low Study & High Sleep:", avg_low_study_high_sleep)
```
